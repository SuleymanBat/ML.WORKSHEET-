# Write a Python program to create a plot to get a general Statistics of Iris data.

```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("../input/irisdataset/Iris.csv")
iris.describe().plot(kind = "area",fontsize=16, figsize = (15,8), table = True, colormap="Accent")
plt.xlabel('Statistics',)
plt.ylabel('Value')
plt.title("General Statistics of Iris Dataset")
plt.show()
```

# Write a Python program to create a Bar plot to get the frequency of the three species of the Iris data

```python
import pandas as pd
import seaborn as sns
iris = pd.read_csv("../input/irisdataset/Iris.csv")
ax = plt.subplots(1,1,figsize = (10,8))
sns.countplot('iris', data = iris)
plt.title('IRIS SPECIES COUNT')
plt.show()
```

# Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data.

```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv('../input/irisdataset/Iris.csv')
ax = plt.subplots(1,1,figsize = (10,8))
iris['iris'].value_counts().plot.pie(explode=[0.1,0.1,0.1], autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("IRIS SPECIES %")
plt.show()
```

# Write a Python program to create a graph to find relationship between the sepal length and width.

```python
fig = iris[iris.iris=='Iris-setosa'].plot(kind='scatter',x='sepal-length',y='sepal-width',color='orange', label='Setosa')
iris[iris.iris=='Iris-versicolor'].plot(kind='scatter',x='sepal-length',y='sepal-width',color='blue', label='versicolor',ax=fig)
iris[iris.iris=='Iris-virginica'].plot(kind='scatter',x='sepal-length',y='sepal-width',color='green', label='virginica', ax=fig)
fig.set_xlabel("sepal-length")
fig.set_ylabel("sepal-width")
fig.set_title("sepal-length VS width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```
